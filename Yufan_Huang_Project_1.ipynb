```python
clearscreen()
#set up canvas
setup(200,200)
bgcolor('black')

#draw white rectangles
cat = Turtle()
cat.color((245,244,242))

cat.jumpto(-100,400)
cat.begin_fill()
cat.forward(200)
cat.right(90)
cat.forward(300)
cat.left(90)
cat.forward(20)
cat.right(90)
cat.forward(40)
cat.right(90)
cat.forward(20)
cat.left(90)
cat.forward(40)
cat.left(90)
cat.forward(60)
cat.right(90)
cat.forward(20)
cat.right(90)
cat.forward(260)
cat.right(90)
cat.forward(170)
cat.left(90)
cat.forward(100)
cat.right(90)
cat.forward(20)
cat.right(90)
cat.forward(100)
cat.left(90)
cat.forward(200)
cat.end_fill()

#draw white dashed lines
cat.jumpto(-100,60)
cat.pensize(10)
cat.right(90)
cat.backward(100)
cat.penup()
cat.backward(14)
cat.pendown()
cat.backward(60)
cat.penup()
cat.backward(14)
cat.pendown()
cat.backward(30)
cat.pensize(1)

#draw repeated pattern
cat.jumpto(-96,-50)
def repeated_pattern(turtle_obj, side_length):
  for i in range(7):
    turtle_obj.forward(side_length)
    turtle_obj.penup()
    turtle_obj.forward(20)
    turtle_obj.pendown()
repeated_pattern(cat, 16)
cat.jumpto(-88,-10)
def repeated_pattern_2(turtle_obj, side_length):
  for i in range(7):
    turtle_obj.right(90)
    turtle_obj.pendown()
    turtle_obj.forward(side_length)
    turtle_obj.penup()
    turtle_obj.left(90)
    turtle_obj.forward(36)
    turtle_obj.left(90)
    turtle_obj.forward(side_length)
    turtle_obj.right(90)
    turtle_obj.pendown()
repeated_pattern_2(cat, 40)

#draw filled squares
cat.jumpto(140,100)
def filled_square(turtle_obj, side_length_1, side_length_2):
  for i in range(1):
    turtle_obj.begin_fill()
    turtle_obj.forward(side_length_1)
    turtle_obj.right(90)
    turtle_obj.forward(side_length_2)
    turtle_obj.right(90)
    turtle_obj.forward(side_length_1)
    turtle_obj.right(90)
    turtle_obj.forward(side_length_2)
    turtle_obj.end_fill()
filled_square(cat,90,40)
cat.jumpto(-80,-200)
filled_square(cat,20,30)
cat.jumpto(-180,-200)
filled_square(cat,100,30)
dog = Turtle()
dog.jumpto(-60,-100)
dog.color("orange")
filled_square(dog,100,10)
rabbit = Turtle()
rabbit.color("purple")
rabbit.jumpto(-300,-200)
filled_square(rabbit,80,30)
cat.jumpto(-330,-160)
filled_square(cat,30,40)
cat.jumpto(-270,-240)
filled_square(cat,30,40)

#draw circles
def draw_polygon(turtle_obj, side_number, side_length):
  for side in range(side_number):
    turtle_obj.fd(side_length)
    turtle_obj.lt(360/side_number)
def draw_circle(turtle_obj, radius):
  from math import pi as pi
  circumference = 2*pi*radius
  sides = int(circumference/3)+3
  length = circumference/sides
  draw_polygon(turtle_obj, sides, length)
rabbit.jumpto(160,80)
rabbit.begin_fill()
draw_circle(rabbit, 5)
rabbit.end_fill()
rabbit.jumpto(-100,0)
draw_circle(rabbit, 60)
dog.jumpto(500,-200)
dog.begin_fill()
draw_circle(dog, 150)
dog.end_fill()
```
