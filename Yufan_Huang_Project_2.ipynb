```python
clearscreen()
t=Turtle()
speed(100)
setup(500,500)
bgcolor((255,255,255))

#create random shapes
for i in range (1,19):
  r=random.randint(0,255)
  g=random.randint(0,255)
  b=random.randint(0,255)
  t.fillcolor((r,g,b))
  t.color((r,g,b))
  x = random.randint(-200,200)
  y = random.randint(-200,200)
  side_length_1 = x-i
  side_length_2 = y+i
  t.begin_fill()
  t.forward(side_length_1)
  t.right(90)
  t.forward(side_length_2)
  t.right(90)
  t.forward(side_length_1)
  t.right(90)
  t.forward(side_length_2)
  t.end_fill()
  t.penup()
  t.goto(x,y)
  t.pendown()

def draw_polygon(t, side_number, side_length):
  for side in range(side_number):
    t.fd(side_length)
    t.lt(360/side_number)
def draw_circle(t, radius):
  from math import pi as pi
  circumference = 2*pi*radius
  sides = int(circumference/3)+3
  length = circumference/sides
  draw_polygon(t, sides, length)
for i in range (20,30):
  r=random.randint(0,255)
  g=random.randint(0,255)
  b=random.randint(0,255)
  t.color((r,g,b))
  x = random.randint(-200,200)
  y = random.randint(-200,200)
  draw_circle(t, i)
  t.penup()
  t.goto(x,y)
  t.pendown()
```

```python
screen_obj = Screen()
screen_obj.saveSVG('random 1.svg', False)
```
